```python
import numpy as np
import matplotlib.pyplot as plt

# Generar seal senoidal
fs = 1000  # frecuencia de muestreo
t = np.arange(0, 1, 1/fs)  # vector de tiempo
f = 100  # frecuencia de la seal
x = np.sin(2 * np.pi * f * t)  # Definir la variable x para la seal senoidal

# Aplicar Transformada de Fourier
xf = np.fft.fft(x)

# Generar filtro pasa-bajo
n = len(x)
fcutoff = 50  # frecuencia de corte
h = np.ones(n)  # vector de unos
h[int(np.round(n * fcutoff / fs)):] = 0  # aplicar filtro pasa-bajo

# Aplicar filtro a la seal en el dominio de la frecuencia
xf_filtered = xf * h

# Convertir seal filtrada al dominio del tiempo
x_filtered = np.fft.ifft(xf_filtered)

# Graficar seal original y seal filtrada
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.title('Seal original')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

plt.subplot(2, 1, 2)
plt.plot(t, np.real(x_filtered))
plt.title('Seal filtrada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```
