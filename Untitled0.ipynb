```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

n = np.arange(-3, 8)
x = 0.55**(n + 3)
h = np.ones(11)  # Se define una secuencia de impulso discreto
y = convolve(x, h, mode='full')

# Crear tres subgrficas en una figura
plt.figure(figsize=(10, 8))

# Subgrfica 1
plt.subplot(3, 1, 1)
plt.stem(n, x)
plt.title('Seal Original')
plt.grid(True)

# Subgrfica 2
plt.subplot(3, 1, 2)
plt.stem(n, h)
plt.title('Respuesta al impulso/Segunda Seal')
plt.grid(True)

# Subgrfica 3
plt.subplot(3, 1, 3)
plt.stem(np.arange(len(y)), y)
plt.title('Convolucin Resultante')
plt.grid(True)

plt.tight_layout()  # Ajusta el diseo para evitar solapamientos
plt.show()
```
